```python
# 1. Implement custom `enumerate`
# Mô tả: Cài đặt một generator function hoạt động tương tự như hàm `enumerate` có sẵn của Python.

def custom_enumerate(iterable, start=0):
    current_index = start
    for element in iterable:
        yield(current_index,element)
        current_index +=1

def test():
    assert list(custom_enumerate(['a', 'b', 'c'])) == [(0, 'a'), (1, 'b'), (2, 'c')]
    assert list(custom_enumerate('abc', 1)) == [(1, 'a'), (2, 'b'), (3, 'c')]
    assert list(custom_enumerate([], 5)) == []
    print('All tests passed!')

test()
```

```python
# 2. Implement custom `map/filter/reduce`
# Mô tả: Tự viết các hàm mô phỏng lại cách hoạt động của `map`, `filter`, và `reduce`.

def custom_map(func, iterable):
    for element in iterable: 
        yield func(element)

def custom_filter(func, iterable):
    for element in iterable:
        if func(element):
            yield element

def custom_reduce(func, iterable, initial=None):
    it = iter(iterable)
    if initial is None:
        try:
            accumulator = next(it)
        except StopIteration:
            raise TypeError("custom_reduce() of empty sequence with no initial value") 
    else:
        accumulator = initial

    for element in it:
        accumulator = func(accumulator, element)
    
    return accumulator

def test():
    assert list(custom_map(lambda x: x*2, [1, 2, 3])) == [2, 4, 6]
    assert list(custom_filter(lambda x: x%2==0, [1, 2, 3, 4])) == [2, 4]
    assert custom_reduce(lambda x, y: x+y, [1, 2, 3, 4]) == 10
    assert custom_reduce(lambda x, y: x+y, [1, 2, 3, 4], 10) == 20
    print('All tests passed!')

test()
```

```python
# 3. Shallow copy vs deep copy
# Mô tả: Viết một hàm để deep copy (sao chép sâu) một mảng 2 chiều (nested list) mà không dùng thư viện `copy`.

def deep_copy_list(lst):
    if isinstance(lst, list):
        return [deep_copy_list(element) for element in lst]
    return lst

def test():
    original = [[1, 2], [3, 4]]
    copied = deep_copy_list(original)
    copied[0][0] = 99
    assert original[0][0] == 1
    assert copied[0][0] == 99
    print('All tests passed!')

test()
```

```python
# 5. Decorator logging function runtime
# Mô tả: Viết một decorator dùng để in ra thời gian chạy của hàm được trang trí (ví dụ: dùng thư viện `time`).

import time

def timeit(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()

        result = func(*args, **kwargs)

        end_time = time.time()
        execution_time = end_time - start_time
        print(f"Hàm '{func.__name__}' chạy mất {execution_time:.4f} giây")
        return result
    return wrapper

def test():
    @timeit
    def sleep_func():
        time.sleep(0.1)
        return 'done'
    
    assert sleep_func() == 'done'
    print('All tests passed! (Check console for time log)')

test()
```

```python
# 6. Generator vs iterator
# Mô tả: Tạo một class hoạt động như một iterator (có `__iter__` và `__next__`) để trả về n số Fibonacci đầu tiên.

class FibIterator:
    def __init__(self, n):
        self.n = n
        self.length = 0
        self.a = 0
        self.b = 1  

    def __iter__(self):
        return self
    def __next__(self):
        if self.length >= self.n:
            raise StopIteration

        result = self.a
        self.a, self.b = self.b, self.a+self.b
        self.length+=1
        return result


def test():
    assert list(FibIterator(5)) == [0, 1, 1, 2, 3]
    assert list(FibIterator(1)) == [0]
    print('All tests passed!')

test()
```

```python
# 7. Context manager (`with`)
# Mô tả: Tạo một Context Manager (sử dụng class có `__enter__` và `__exit__`) để đo thời gian chạy của một khối lệnh nằm trong câu lệnh `with`.

import time
class TimerContextManager:
    def __enter__(self):
        self.start = time.perf_counter()
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        self.end = time.perf_counter()
        duration = self.end - self.start
        print(f"duration: {duration:.4f} seconds")

def test():
    with TimerContextManager() as t:
        time.sleep(0.1)
    print('All tests passed! (Check console for time)')

test()
```

```python
# 9. Classmethod vs staticmethod
# Mô tả: Tạo một class có chứa một `classmethod` để khởi tạo object từ một chuỗi (ví dụ 'John-25') và một `staticmethod` để kiểm tra độ tuổi hợp lệ (>= 18).

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = int(age)
    
    @classmethod
    def from_string(self, data_string):
        name,age= data_string.split('-')
        return self(name,age)
    
    @staticmethod
    def is_adult(age):
        return age >= 18

def test():
    p = Person.from_string('Alice-20')
    assert p.name == 'Alice' and p.age == 20
    assert Person.is_adult(20) == True
    assert Person.is_adult(15) == False
    print('All tests passed!')

test()
```

```python
# 10. Property getter/setter
# Mô tả: Sử dụng decorator `@property` để viết getter và setter cho thuộc tính `temperature`. Setter phải chặn việc đặt nhiệt độ dưới -273 độ (ném ra ValueError).

class Celsius:
    def __init__(self, temperature=0):
        self._temperature = temperature
    
    @property
    def temperature(self):
        return self._temperature
    
    @temperature.setter
    def temperature(self, value):
        if value < -273:
            raise ValueError("Nhiệt độ không thể dưới -273 độ C (độ không tuyệt đối)")
        self._temperature = value


def test():
    c = Celsius(37)
    assert c.temperature == 37
    c.temperature = 100
    assert c.temperature == 100
    try:
        c.temperature = -300
        assert False
    except ValueError:
        pass
    print('All tests passed!')

test()
```

```python
# 11. Multiple inheritance + MRO
# Mô tả: Cho 3 class A, B, C (C kế thừa A và B). Ghi đè phương thức `speak` trong tất cả các class và sử dụng `super()` để thấy thứ tự MRO (Method Resolution Order). Hàm trả về danh sách các string khi chạy.

class A:
    def speak(self, res):
        res.append('A')
class B:
    def speak(self, res):
        res.append('B')
class C(A, B):
    def speak(self, res):
        res.append('C')
        super().speak(res)
        

def test():
    res = []
    c = C()
    c.speak(res)
    # Thay đổi C hoặc thứ tự gọi super nếu cần, nhưng test này kỳ vọng MRO chuẩn của Python
    assert res == ['C', 'A'], f'Got {res}'
    print('All tests passed!')

test()
```

```python
# 12. Metaclass basic usage
# Mô tả: Tạo một metaclass có nhiệm vụ tự động in hoa (uppercase) tất cả các thuộc tính của class (ngoại trừ các dunder methods bắt đầu bằng __).

class UppercaseMeta(type):
    def __new__(cls, name, bases, dct):
        uppercase_dct = {}
        for key, value in dct.items():
            if not key.startswith('__'):
                uppercase_dct[key.upper()] = value
            else:
                uppercase_dct[key] = value
        return super().__new__(cls, name, bases, uppercase_dct)

class MyClass(metaclass=UppercaseMeta):
    my_var = 10

def test():
    assert hasattr(MyClass, 'MY_VAR')
    assert not hasattr(MyClass, 'my_var')
    assert MyClass.MY_VAR == 10
    print('All tests passed!')

test()
```

```python
# 14. LRU Cache decorator
# Mô tả: Sử dụng decorator `functools.lru_cache` hoặc tự viết memoization decorator cho một hàm tính giai thừa. Ở đây hãy tự viết một decorator `memoize` đơn giản lưu kết quả hàm.

def memoize(func):
    cache = {}

    def wrapper(*args):
        if args in cache:
            return cache[args]
        result = func(*args)
        cache[args] = result
        return result
    return wrapper

@memoize
def expensive_func(n):
    return n * n

def test():
    assert expensive_func(5) == 25
    assert expensive_func(5) == 25 # Should hit cache
    print('All tests passed!')

test()
```

```python
# 15. Thread-safe singleton
# Mô tả: Cài đặt pattern Singleton bằng metaclass sao cho an toàn với đa luồng (thread-safe), sử dụng `threading.Lock`.

import threading

class SingletonMeta(type):
    _instances = {}
    _lock = threading.Lock()

    def __call__(cls, *args, **kwargs):
        if cls not in cls._instances:
            with cls._lock:
                if cls not in cls._instances:
                    instance = super().__call__(*args, **kwargs)
                    cls._instances[cls] = instance
        return cls._instances[cls]

class Singleton(metaclass=SingletonMeta):
    pass

def test():
    s1 = Singleton()
    s2 = Singleton()
    assert s1 is s2
    print('All tests passed!')

test()
```

```python
# 16. Longest substring without repeating chars
# Mô tả: Cho chuỗi s, tìm chiều dài của chuỗi con dài nhất không chứa ký tự lặp lại.

def length_of_longest_substring(s: str) -> int:
    char_index_map = {}
    left = 0
    max_length = 0

    for right in range(len(s)):
        char = s[right]

        # 1. Khối lệnh if: CHỈ có tác dụng dịch chuyển con trỏ left nếu có trùng lặp
        if char in char_index_map and char_index_map[char] >= left:
            left = char_index_map[char] + 1  # Đã sửa lại lỗi +1 dictionary

        # 2. Khối lệnh cập nhật: BẮT BUỘC phải nằm NGOÀI lệnh if (lùi lề ra bằng với chữ if)
        # Vì dù có bị trùng hay không thì ta vẫn phải luôn cập nhật độ dài cửa sổ ở mỗi bước chạy
        char_index_map[char] = right
        
        current_window_length = right - left + 1
        max_length = max(max_length, current_window_length)
    
    return max_length

def test():
    assert length_of_longest_substring('abcabcbb') == 3
    assert length_of_longest_substring('bbbbb') == 1
    assert length_of_longest_substring('pwwkew') == 3
    assert length_of_longest_substring('') == 0
    print('All tests passed!')

test()
```

```python
# 17. Valid palindrome
# Mô tả: Kiểm tra xem một chuỗi có phải là chuỗi đối xứng (palindrome) không, chỉ xét các ký tự chữ cái, số và bỏ qua hoa thường.

def is_palindrome(s: str) -> bool:
    norm_str = "".join(char for char in s if char.isalpha()).lower()
    return norm_str == norm_str[::-1]

def test():
    assert is_palindrome('A man, a plan, a canal: Panama') == True
    assert is_palindrome('race a car') == False
    assert is_palindrome(' ') == True
    print('All tests passed!')

test()
```

```python
# 18. Group anagrams
# Mô tả: Cho một mảng các chuỗi, hãy nhóm các anagrams (các từ tạo ra bằng cách đảo chữ cái) lại với nhau.
from collections import defaultdict

def group_anagrams(strs: list[str]) -> list[list[str]]:
    anagram_map = defaultdict(list)
    for s in strs:
        sorted_key = "".join(sorted(s))
        anagram_map[sorted_key].append(s)
    return list(anagram_map.values())

def test():
    res = group_anagrams(['eat','tea','tan','ate','nat','bat'])
    res = [sorted(g) for g in res]
    assert sorted(res) == sorted([['ate','eat','tea'], ['bat'], ['nat','tan']])
    print('All tests passed!')

test()
```
