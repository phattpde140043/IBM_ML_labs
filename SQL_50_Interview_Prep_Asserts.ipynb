Mỗi bài tập sẽ có hàm `test_xx()`. Hàm này sẽ tự động chạy query của bạn, chạy một câu query "đáp án chuẩn", và dùng Pandas để so sánh (assert) hai kết quả. Nếu mọi thứ khớp nhau, nó sẽ in ra **All tests passed! 🚀**



```python
import sqlite3
import pandas as pd
from IPython.display import display

# Tạo in-memory database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

cursor.executescript('''
CREATE TABLE users (user_id INTEGER PRIMARY KEY, name TEXT, age INTEGER, email TEXT, country TEXT, salary REAL, department_id INTEGER);
CREATE TABLE orders (order_id INTEGER PRIMARY KEY, user_id INTEGER, amount REAL, order_date DATE);
CREATE TABLE departments (department_id INTEGER PRIMARY KEY, department_name TEXT);
CREATE TABLE employees (employee_id INTEGER PRIMARY KEY, name TEXT, salary REAL, department_id INTEGER, manager_id INTEGER);

INSERT INTO departments VALUES (1, 'IT'), (2, 'HR'), (3, 'Sales'), (4, 'Marketing');
INSERT INTO users VALUES 
(1, 'Alice', 25, 'alice@gmail.com', 'USA', 6000, 1),
(2, 'Bob', 30, 'bob@yahoo.com', 'UK', 4500, 2),
(3, 'Charlie', 22, 'charlie@gmail.com', 'USA', 7000, 1),
(4, 'David', 35, 'david@hotmail.com', 'Canada', 3000, 3),
(5, 'Eve', 28, 'eve@gmail.com', 'UK', 5500, 1),
(6, 'Frank', 40, 'frank@company.com', 'Australia', 8000, 3),
(7, 'Grace', 27, 'grace@gmail.com', 'Canada', 4800, 2),
(8, 'Alice', 29, 'alice.duplicate@gmail.com', 'USA', 5200, 4);

INSERT INTO orders VALUES
(101, 1, 250.0, '2023-01-15'), (102, 1, 150.0, '2023-02-20'),
(103, 2, 500.0, '2023-01-18'), (104, 3, 1000.0, '2023-03-05'),
(105, 5, 300.0, '2023-02-10'), (106, 6, 800.0, '2023-04-12');

INSERT INTO employees VALUES
(1, 'Big Boss', 10000, 1, NULL), (2, 'Manager A', 8000, 1, 1),
(3, 'Employee 1', 5000, 1, 2), (4, 'Employee 2', 6000, 1, 2),
(5, 'Manager B', 7500, 2, 1), (6, 'Employee 3', 4000, 2, 5);
''')
conn.commit()

def check_sql_result(user_query, expected_query, conn):
    user_df = pd.read_sql_query(user_query, conn)
    expected_df = pd.read_sql_query(expected_query, conn)
    
    # Assert
    assert len(user_df) == len(expected_df), f"Số dòng không khớp: User {len(user_df)} != Expected {len(expected_df)}"
    assert len(user_df.columns) == len(expected_df.columns), f"Số cột không khớp: User {len(user_df.columns)} != Expected {len(expected_df.columns)}"
    
    # So sánh values
    assert user_df.values.tolist() == expected_df.values.tolist(), "Dữ liệu trả về chưa chính xác (Lưu ý: thứ tự cột và dòng cũng phải khớp)"
    
    display(user_df.head())
    print("✅ All tests passed! 🚀")

print("Database and Mock Data initialized successfully!")

```
